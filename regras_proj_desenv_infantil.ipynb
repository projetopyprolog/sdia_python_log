```python
def consulta_marcos_presentes_para_crianca(crianca_nome,marcos_lista):
  lista_retorno_consulta = []

  for m in marcos_lista:
    consulta = f"{m}_presente({crianca_nome})"
    #print(consulta)
    resultado_consulta = base_conhecimento.query(pl.Expr(consulta))
    #print(resultado_consulta)
    #postura_presente(Cod,C) :- postura(Cod,C,presente,Reg_idade_marco, Grupo_mes_limite), Reg_idade_marco=< Grupo_mes_limite, criança(Cod,C,Idade), Idade>=Reg_idade_marco
    lista_retorno_consulta.append(resultado_consulta[0])

  return lista_retorno_consulta

def dev_adequado_crianca(crianca_nome,marcos_lista):
  resultados_marcos_lista = consulta_marcos_presentes_para_crianca(crianca_nome, marcos_lista)
  #print(*resultados_marcos_lista, sep='\n')

  classificacao = False
  for r_mrc in resultados_marcos_lista:
    #print(r_mrc)
    if r_mrc == "No":
      classificacao = False
      break
    else:
      classificacao = True

  return classificacao
```
